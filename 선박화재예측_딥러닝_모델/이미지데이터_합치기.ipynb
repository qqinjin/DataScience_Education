```python
# vaild
import os
a = 0
b = 0
c = 0
# 입력 폴더 경로, 출력 폴더 경로 설정
input_folder_path = "/content/test_f/labels/"  # 입력 폴더 경로
output_folder_path = "/content/test_valid/"  # 출력 폴더 경로

# 출력 폴더가 존재하지 않으면 새 폴더 생성
if not os.path.exists(output_folder_path):
    os.makedirs(output_folder_path)

# 입력 폴더 내 모든 파일 검색
for filename in os.listdir(input_folder_path):
    # 파일이 텍스트 파일인 경우에만 작업 수행
    if filename.endswith(".txt"):
        # 파일 내용 읽어오기
        with open(os.path.join(input_folder_path, filename), "r") as f:
            content = f.read()

        # 파일 내용을 개행문자를 기준으로 분리하여 처리
        lines = content.split("\n")
        for i, line in enumerate(lines):
            # 각 라인 내 첫 숫자가 0인 경우 2로, 1인 경우 3으로 변경
            if line.startswith("0"):
                a += 1
            elif line.startswith("1"):
                b += 1
            elif line.startswith("2"):
                c += 1

        # 변경된 라인을 다시 문자열로 결합하여 새 파일에 쓰기
        new_file_path = os.path.join(output_folder_path, filename)
        with open(new_file_path, "w") as f:
            f.write("\n".join(lines))
```

# train

```python
import os

images_dir = "/content/all_data_go/train/images"
labels_dir = "/content/all_data_go/train/labels"

for image_file in os.listdir(images_dir):
    if image_file.endswith(".jpg"):
        label_file = os.path.join(labels_dir, os.path.splitext(image_file)[0] + ".txt")
        if not os.path.exists(label_file):
            print("No label file found for image: ", os.path.join(images_dir, image_file))
            os.remove(os.path.join(images_dir, image_file))
```

```python
# 기존 폴더에 넣은 이미지,라벨링 개수 확인
train_image = glob('/content/all_data_go/train/images/*.jpg')
train_labels = glob('/content/all_data_go/train/labels/*.txt')

print(len(train_image))
print(len(train_labels))
```

```python
import shutil
import os
from google.colab import files

# 압축할 폴더 경로
folder_path = "/content/all_data_go"

# 압축 파일 이름
zip_file_name = "all_data_go"

# 압축 파일 경로
zip_file_path = os.path.join(os.getcwd(), zip_file_name)

# shutil 모듈의 make_archive 함수를 사용하여 폴더를 압축합니다.
shutil.make_archive(zip_file_path, 'zip', folder_path)

# 파일 다운로드
files.download(zip_file_path)
```
