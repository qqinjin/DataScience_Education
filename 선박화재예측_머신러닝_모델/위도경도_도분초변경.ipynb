```python
import numpy as np
import pandas as pd
ship_fire = pd.read_csv('/content/drive/MyDrive/deu2/4월_프로젝트/Machine_dataset/all_ship_fire_data.csv', encoding = 'cp949')
```

```python
ship_fire
```

```python
import pandas as pd

# 위도 컬럼 선택 후 도, 분, 초 분리
latitudes = ship_fire['위도'].str.split('|', expand=True).astype(float)


# for문을 사용하여 모든 위도 값들에 적용
for i in range(len(ship_fire['위도'])):
    lat = latitudes.iloc[i]
    lat_minute = (lat[1] + lat[2] / 60) / 60
    lat_degree = lat[0] + lat_minute
    ship_fire['위도'][i] = lat_degree

# 결과 확인
ship_fire
```
