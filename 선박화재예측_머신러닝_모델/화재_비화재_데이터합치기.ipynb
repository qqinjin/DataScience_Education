```python
from geopy.distance import geodesic

# first_df에 지점 컬럼 추가
all_df['지점'] = None

# first_df와 df의 위도, 경도 비교하여 가장 가까운 df의 행 선택
for i, row in all_df.iterrows():
    if row['위도'] < -90 or row['위도'] > 90:
        continue  # 범위를 벗어나는 경우 다음 행으로 이동
    min_dist = float('inf')
    closest_row = None
    for j, df_row in df.iterrows():
        dist = geodesic((row['위도'], row['경도']), (df_row['위도'], df_row['경도'])).km
        if dist < min_dist:
            min_dist = dist
            closest_row = df_row
    all_df.at[i, '지점'] = closest_row['지점']
```

```python
# first_df와 second_df를 key값을 기준으로 merge하여 병합
merged_df = pd.merge(all_df, second_df, on='key')

# 결과 데이터프레임 초기화
result_df = pd.DataFrame(columns=list(merged_df.columns))

# merged_df의 행을 하나씩 처리
for index, row in merged_df.iterrows():
    # key 값 추출
    key_value = row['key']

    # second_df에서 key 값이 일치하는 행 추출
    filtered_rows = second_df.loc[second_df['key'] == key_value]

    # second_df에서 추출한 행들을 result_df에 이어붙이기
    for _, filtered_row in filtered_rows.iterrows():
        # 첫번째 데이터프레임과 두번째 데이터프레임의 컬럼명이 겹치면 에러가 발생하므로 첫번째 데이터프레임 컬럼명을 변경합니다.
        filtered_row.index = [f'{col}_second' for col in filtered_row.index]

        # result_df에 이어붙이기
        result_row = pd.concat([row, filtered_row[1:]], axis=0)
        result_df = result_df.append(result_row, ignore_index=True)

result_df
```
